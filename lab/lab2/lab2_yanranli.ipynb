Q4

```python
size = 1000000
rv1 = np.random.normal(0, 1, size)
rv2 = np.random.normal(0, 1, size)
rv_sum = rv1 + rv2
rv_pro = rv1 * rv2

plt.subplot(1,2,1)
plt.hist(rv_sum, density= True, bins = 30, rwidth = 0.95)
plt.title('Sum of two rvs')
plt.subplot(1,2,2)
plt.hist(rv_pro, bins = 30, rwidth = 0.95, density= True)
plt.title('Product of two rvs')
plt.show()
```
