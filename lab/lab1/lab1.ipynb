```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
G = nx.Graph()
for node in range(1,21):
    G.add_edge(0,node)
plt.figure()
nx.draw(G)
plt.savefig("lab1_fig1.pdf")
```

```python
x = np.linspace(-20,20,100)
y = np.sin (x) / x
plt.figure()
plt.plot(x,y,".-")
plt.savefig("lab1_fig2.pdf")
```
