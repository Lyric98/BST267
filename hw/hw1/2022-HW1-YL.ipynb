```python
# RUN THIS CODE
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib notebook

# generate a random graph realization
G = nx.erdos_renyi_graph(30, 0.06, seed=2022)

# visualize the graph
plt.figure()
nx.draw(G, node_size=50)
```

**ANSWERS:**

List the size of each component from largest to smallest: 19, 4, 3, 2, 1, 1

(b) If we select one of the nodes in this graph uniformly at random, the probability that the selected node belongs to the largest connected component is 19/(4+3+2+1+1+19) = 0.633
